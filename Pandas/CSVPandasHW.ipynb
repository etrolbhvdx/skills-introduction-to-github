## **Task 1: Read into CSV**

```python
import pandas as pd

acc_df=pd.read_csv('acc_players-2324F.csv',header=1)
acc_df.head()
```

## **Task 2: Basic Analysis**

```python
#Calculate total points scored
print('Total points scored by all players:')
total_points=acc_df['PTS'].sum()
print(total_points)
#Find the player with the most minutes played
print('Player with most minutes played:')
minutes=acc_df[['Player','MP']].sort_values(by='MP',ascending=False).head(1)
print(minutes)
#Find the top 5 players by total rebounds
print('Top 5 players by total rebounds:')
rebounders=acc_df[['Player','TRB']].sort_values(by='TRB',ascending=False).head(5)
print(rebounders)
```

## **Task 3: Player Filtering**

```python
#Create a dataframe only with players with more than 500 minutes played
filtered_df=acc_df[acc_df['MP']>500]
#Determine player with highest total assists
print('Player with highest total assists:')
assists=filtered_df[['Player','AST']].sort_values(by='AST',ascending=False).head(1)
print(assists)
#Find top three assist leaders
print('Top three assist leaders:')
assist_leaders=filtered_df[['Player','AST']].sort_values(by='AST',ascending=False).head(3)
print(assist_leaders)
#Find top three shot blockers:
print('Top three shot blockers:')
shot_blockers=filtered_df[['Player','BLK']].sort_values(by='BLK',ascending=False).head(3)
print(shot_blockers)
```

## **Task 4: School Based Analysis**

```python
#Group players by school and calculate total points scored by each school
print('Total points scored by each school:')
school_points=acc_df.groupby('School')['PTS'].sum()
print(school_points)
#Group players by school and calculate total assists scored by each school
print('Total assists scored by each school:')
school_assists=acc_df.groupby('School')['AST'].sum()
print(school_assists)
#Sort schools by total points scored and display the top three schools
print('Top three schools by total points scored:')
top_schools=school_points.sort_values(ascending=False).head(3)
print(top_schools)
```

## **Task 5: Extra Credit**

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Create a bar chart showing the top five players by total points scored
top_players_points=acc_df[['Player','PTS']].sort_values(by='PTS',ascending=False).head(5)
print(top_players_points)
plt.figure(figsize=(10,6))
sns.barplot(x='PTS', y='Player', data=top_players_points)
plt.title('Five Highest Scoring Players')
plt.xlabel('Total Points')
plt.show()
```

```python
#Investigate whether there is any correlation between a player's field goal percentage (FG%) and their win shares (WS).
#Provide a scatter plot and calculate the correlation coefficient.
plt.figure(figsize=(10,6))
sns.scatterplot(x='FG%', y='WS', data=acc_df)
plt.title('Field Goal Percentage vs Win Shares')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Win Shares')
plt.show()
corr_coefficient=acc_df['FG%'].corr(acc_df['WS'])
print('Correlation coefficient:', corr_coefficient)
```

The correlation coefficient is 0.421. This is positive, which means based on this dataset, win shares generally tend to increase when a player's FG% increases. Generally, scoring more points is better than fewer, so some positive correlation is to be expected. However, 0.421 is not high enough to suggest the variables are strongly correlated, meaning there are probably other variables that are more closely connected with win shares.
